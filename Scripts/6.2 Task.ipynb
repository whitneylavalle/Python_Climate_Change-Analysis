# 04. correlation matrix heatmap (colored)

```python
sub=df[['Coal Consumption - EJ', 'Gas Consumption - EJ',
       'Geo Biomass Other - TWh', 'Hydro Generation - TWh',
       'Nuclear Generation - TWh', 'Solar Generation - TWh',
       'Wind Generation -TWh', 'Oil Consumption - EJ']]
```

```python
sub.head()
```

```python
# create a subplot with matplotlib

f,ax = plt.subplots(figsize=(10,10))
```

```python
# create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 

# the `annot` argument allows the plot to place the correlation coefficients onto the heatmap.
```

```python
corr.figure
```

```python
# the plot shows the relation between fossil fuel usage, oil/gas/coal are all correlated strongly to one another
# the plot also shows a weaker relationship between solar power consumption and oil/nuclear/gas/coal consumption
# the plot also shows a linkage between renewable consumption (e.g. wind and solar)
# these trends make sense as the consumption of fossil fuels and renewables represent polar opposite ends of the spectrum of sustainability and public support
```

# 05. scatterplots for variables with strongest relationships

```python
sub.columns
```

```python
# r=0.96
sns.lmplot(x = 'Gas Consumption - EJ', y = 'Oil Consumption - EJ', data = df)
```

```python
# r=0.92
sns.lmplot(x = 'Gas Consumption - EJ', y = 'Nuclear Generation - TWh', data = df)
```

```python
# r=0.93
sns.lmplot(x = 'Solar Generation - TWh', y = 'Wind Generation -TWh', data = df)
```

```python
# this relationship is surprising (r=0.95) and I would liket o investigate throughout this analysis
sns.lmplot(x = 'Oil Consumption - EJ', y = 'Hydro Generation - TWh', data = df)
```

### The first 3 show the results found with the colored correlation heatmap matrix, that non-renewables are highly correlated and renewables are highly correlated. This is somewhat expected as investment in solar power for example would point to a likely investment in renewables as a whole. Likewise, a national reliance on oil would indicate more reliance across the board on fossil fuels. 

### The curiously strong relationship is between hydrogeneration and oil consumption. Many environmental organizations deem hydropower to be one of the most harmful "renewable" practices to the environment, leaving it surprisingly correlated with non-renewables, especially oil.
